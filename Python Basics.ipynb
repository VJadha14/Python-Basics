```python
# 6. Write a function to demonstrate the behavior of mutable and immutable arguments ?

def demonstrate_mutable_immutable(immutable_arg, mutable_arg):

    print(f"Immutable before: {immutable_arg}")
    immutable_arg += 1  #
    print(f"Immutable after: {immutable_arg}")

    print(f"Mutable before: {mutable_arg}")
    mutable_arg.append(4)  #
    print(f"Mutable after: {mutable_arg}")

immutable_value = 10

mutable_list = [1, 2, 3]

demonstrate_mutable_immutable(immutable_value, mutable_list)


print(f"Immutable outside: {immutable_value}")
print(f"Mutable outside: {mutable_list}")
```

```python
# 15 Write a Python program to calculate the factorial of a number provided by the user using a while loop?
def demonstrate_mutable_immutable(mutable_arg, immutable_arg):
    print("Original mutable argument:", mutable_arg)
    print("Original immutable argument:", immutable_arg)

    # Modify the mutable argument (list)
    mutable_arg.append(100)  # Add a new element to the list

    # Modify the immutable argument (integer)
    immutable_arg += 10  # This doesn't change the original value, because integers are immutable

    print("Modified mutable argument:", mutable_arg)
    print("Modified immutable argument:", immutable_arg)

# Example usage
my_list = [1, 2, 3]
my_int = 5

# Call the function with a mutable and an immutable argument
demonstrate_mutable_immutable(my_list, my_int)

# After function call, check the values
print("\nAfter function call:")
print("Mutable list outside function:", my_list)
print("Immutable integer outside function:", my_int)
```
